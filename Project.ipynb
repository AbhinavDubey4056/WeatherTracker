```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("weatherHistory.csv")
df.head()
```

```python

df = df.dropna()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
```

```python
df = df.dropna()
```

```python
print(df.dtypes)
```

```python
df.head()
```

```python
print("max temp:",df['Temperature (C)'].max())
print("min temp:",df['Temperature (C)'].min())
print("avg temp:",df["Temperature (C)"].mean())

```

```python
df['Humidity'] = df['Humidity']*100
```

```python
plt.figure(figsize=(14,6))
plt.plot(df["Formatted Date"],df["Temperature (C)"], color='orange',linewidth=0.5)
plt.xlabel('Formatted Date')
plt.ylabel('Temperature (C')
plt.title("temperature trend over time")
plt.grid(True)
plt.show()
```

```python
df['Formatted Date'] = pd.to_datetime(df["Formatted Date"], utc=True)




df['month'] = df['Formatted Date'].dt.month

monthly_avg = df.groupby('month')['Temperature (C)'].mean()
```

```python
plt.figure(figsize=(10,5))
monthly_avg.plot(kind='bar',color="blue")
plt.xticks(ticks=range(12),rotation=0)
plt.xlabel('month')
plt.ylabel('Temperature (C)')
plt.title("Monthly avg temp")
plt.grid(axis = 'y')
plt.tight_layout()
plt.show()
```
